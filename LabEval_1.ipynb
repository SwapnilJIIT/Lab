```python

student_database = {}


def init():
    global student_database
    student_database = {
        '1' : {
            'name' : "Swapnil",
            'std' : '12',
            'grades' : [20]
        },
        '2' : {
            'name' : "Harsh",
            'std' : '11',
            'grades' : [10,15,10]
        },
        '3' : {
            'name' : "Vansh",
            'std' : '12',
            'grades' : [10,25,18]
        },
        '4' : {
            'name' : "Rahul",
            'std' : '11',
            'grades' : [10,12,12]
        }
    }
    
    

def add_student():
    roll = input("Enter student enroll : ")
    name = input("Enter student name : ")
    std = input("Enter student's class : ")
    cnt = int(input("Enter number of grades of student : "))
    grades = []
    for i in range(cnt):
        g = input(f"Enter grade of exam {i+1} : ")
        grades.append(int(g))
        
    student = {}
    student['name'] = str(name)
    student['std'] = str(std)
    student['grades'] = grades
    
    student_database[roll] = student
    
    

    
def calculate_average(r = ""):
    if not r:
        roll = input("Enter student enroll : ")
    else:
        roll = r
    if roll not in student_database:
        print("Student not found")
        return
    avg = 0
    for g in student_database[roll]['grades']:
        avg += g
        
    cnt = len(student_database[roll]['grades'])
    
    avg = avg/cnt
    if not r:
        print("Average of students grade is ", avg)
    else:
        return avg
    
    
    

def update_grades():
    global student_database
    roll = input("Enter student enroll : ")
    if roll not in student_database:
        print("Student not found")
        return
    cnt = int(input("Enter number of grades to be added : "))
    for i in range(cnt):
        g = input(f"Enter grade of exam {i+1} : ")
        student_database[roll]['grades'].append(int(g))
    
    
    
def change_old_grades():
    global student_database
    roll = input("Enter student enroll : ")
    if roll not in student_database:
        print("Student not found")
        return
    indx = int(input("Enter for which exam grades are to be changed : "))
    l = len(student_database[roll]['grades'])
    if indx>l:
        print("Student has not given that exam yet")
        return
    new_g = int(input("Enter new grades : "))
    student_database[roll]['grades'][indx-1] = new_g
    

    
def generate_top_students():
    top_student = {}
    for k,stud in student_database.items():
        if stud['std'] in top_student:
            if calculate_average(top_student[stud['std']][1])<calculate_average(k):
                top_student[stud['std']] = (stud['name'],k)
        else:
            top_student[stud['std']] = (stud['name'],k)
    
    print("TOP PERFORMERS ARE :- ")
    for k,v in top_student.items():
        print(f"Class {k} -> {v[0]}")
        

        
        
init()
# add_student()

# update_grades()
# calculate_average()
# generate_top_students()
# calculate_average()
# change_old_grades()
# calculate_average()









```
